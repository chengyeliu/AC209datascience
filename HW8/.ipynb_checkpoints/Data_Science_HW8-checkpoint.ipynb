```python
#--------  plot_decision_boundary
# A function that visualizes the data and the decision boundaries
# Input: 
#      x (predictors)
#      y (labels)
#      model (classifier)
#      poly_flag (fits quadratic model if true, otherwise linear)
#      title (title for plot)
#      ax (a set of axes to plot on)
# Returns: 
#      ax (axes with data and decision boundaries)

def plot_decision_boundary(x, y, model, title, ax, bounds=(0, 1), poly_flag=False):
    # Plot data
    ax.scatter(x[y == 1, 0], x[y == 1, 1], c='green')
    ax.scatter(x[y == 0, 0], x[y == 0, 1], c='white')
    
    # Create mesh
    interval = np.arange(bounds[0], bounds[1], 0.01)
    n = np.size(interval)
    x1, x2 = np.meshgrid(interval, interval)
    x1 = x1.reshape(-1, 1)
    x2 = x2.reshape(-1, 1)
    xx = np.concatenate((x1, x2), axis=1)

    # Predict on mesh points
    if(poly_flag):
        quad_features = preprocessing.PolynomialFeatures(degree=2)
        xx = quad_features.fit_transform(xx)
        
    yy = model.predict(xx)    
    yy = yy.reshape((n, n))

    # Plot decision surface
    x1 = x1.reshape(n, n)
    x2 = x2.reshape(n, n)
    ax.contourf(x1, x2, yy, alpha=0.1, cmap='Greens')
    
    # Label axes, set title
    ax.set_title(title)
    ax.set_xlabel('Latitude')
    ax.set_ylabel('Longitude')
    
    return ax
```

It is therefore important to select the best support vectors when fitting the model for maximum accuracy. 

## Step 2: Play with values of the parameters of the model

In defining the support vector classifier, we needed to specify the parameter C and kernel. 

C is a tuning paramter which bounds the sum of the errors of each classification. It determines the number and severity of the violations to the margin and hyperplane that should be tolerated. It effectively serves as a budget for the amount that the margin can be violated by the n observations. 

For C > 0 no more than C observations can be on the wrong side of the hyperplane. As C increases the width of the margin increases as the model becomes more tolerant of violations to the margin. 

The value of C also affects the number of support vectors and so also affects the variance of the model (many support vectors means low variance).

```python
#--------  fit_and_plot_svm_for_c
# Fit and plot SVM model for value of 'C', overlayed on a scatter plot of data 
# (fit on train set and evaluate on test set)
#
# Input: 
#      x_train (array of train predictors)
#      y_train (array of train responses)#      
#      x_test (array of test predictors)
#      y_test (array of test responses)
#      bounds (tuple of bounds for plotting)
#      C  (value for parameter C)
#      ax (axes to plot on)

def fit_and_plot_svm_for_c(x_train, y_train, x_test, y_test, C, ax, bounds=(0, 1)):
    # Fit SVM model
    model = svm.SVC(C=C, kernel='linear')
    model.fit(x_train, y_train)
    
    # Train and test error
    tr_acc = model.score(x_train, y_train)
    ts_acc = model.score(x_test, y_test)

    # Plot decision boundary
    plot_decision_boundary(x_train, y_train, model, \
                           'C = ' + str(C)\
                           + ', train acc = ' + str(tr_acc)\
                           + ', test acc = ' + str(ts_acc), ax, bounds)
    
    # Plot support vectors
    sv_indices = model.support_ # retrieve the support vector indices
    ax.scatter(x_train[sv_indices, 0], x_train[sv_indices, 1], color='red', alpha=0.15, s=100) # draw circles around SVs
    
    return ax
```

```python
# Load train data
data_train = np.loadtxt("datasets/dataset_2_train.txt", delimiter=',')
x_train = data_train[:, 0:-1]
y_train = data_train[:, -1]

# Load test data
data_test = np.loadtxt("datasets/dataset_2_test.txt", delimiter=',')
x_test = data_test[:, 0:-1]
y_test = data_test[:, -1]

# Fit and plot for different 'C' values
fig, ax = plt.subplots(1, 3, figsize = (15, 3))

ax[0] = fit_and_plot_svm_for_c(x_train, y_train, x_test, y_test, 0.1, ax[0], (-10, 10))

ax[1] = fit_and_plot_svm_for_c(x_train, y_train, x_test, y_test, 1, ax[1])

ax[2] = fit_and_plot_svm_for_c(x_train, y_train, x_test, y_test, 100, ax[2])

plt.tight_layout()
```

## Analyis

**Are there any obvious draw backs to support vector classifiers as we have presented them to you? What might be some intuitive ways to address these draw backs?**

For `dataset_2`, any linear decision boundary would have some errors. In this case, SVM chooses a decision boundary by trading-off the errors for margin (i.e. balancing minimizing error with maximizing margin). It appears that the parameter $C$ controls the trade-off. 

Increasing $C$ lays more emphasis on accuracy and less emphasis on the margin of separation. As a result, the model tends to overfit the train set, and perform poorly on the test set.

For $C=0.01$, the model clearly underfits the train set. For $C=100$, the model overfits the train set. For $C=1$, the fitted model misclassifies a point in the upper left corner (which could have othwerwise been classified correctly by a linear model), and achieves higher margin.

The support vector classifier uses only linear decision boundaries which perform poorly on datasets in which the classes do not have a clear lineary boundary. 

### Step 3: Non linear decision boundaries

Non linear decision boundaries can be achieved by transforming the feature space in non linear ways e.g using quadratic terms etc. This can be achieved by using kernels. Using a kernel with a support vector classifier results in a support vector machine classifier. The transformation of the feature vector with a kernal essentially amounts to fitting a support vector classifier in a higher-dimensional space. 

```python
def evaluate_kernels(data_train, data_test):
    x_train = data_train[:, 0:-1]
    y_train = data_train[:, -1]
    x_test = data_test[:, 0:-1]
    y_test = data_test[:, -1]

    # Initalise plots
    fig, ax = plt.subplots(1, 3, figsize = (30, 7))

    # Fit SVM model with C = 1000, and polynomial and radial kernels
    for i, kernel in enumerate(['linear', 'poly', 'rbf']):
        model = svm.SVC(C=1000, kernel=kernel)
        model.fit(x_train, y_train)
        # Highlight the support vectors
        sv_indices = model.support_ # retrieve the support vector indices
        ax[i].scatter(x_train[sv_indices, 0], x_train[sv_indices, 1], color='red', alpha=0.15, s=100) # draw circles around SVs
            # Train and test error
        tr_acc = model.score(x_train, y_train)
        ts_acc = model.score(x_test, y_test)
        ax[i] = plot_decision_boundary(x_train, y_train, model, 
                                       str(kernel) + ' kernel'+', train acc = ' + str(tr_acc)\
                               + ', test acc = ' + str(ts_acc), ax[i])
```

```python
#--------  fit_and_plot_dt
# Fit decision tree with on given data set with given depth, and plot the data/model
# Input: 
#      fname (string containing file name)
#      depth (depth of tree)

def fit_and_plot_dt(data_train, data_test, depth, ax):
    x_train = data_train[:, 0:-1]
    y_train = data_train[:, -1]
    x_test = data_test[:, 0:-1]
    y_test = data_test[:, -1]
    # FIT DECISION TREE MODEL
    dt = tree.DecisionTreeClassifier(max_depth = depth)
    dt.fit(x_train, y_train)
    
    # Train and test error
    tr_acc = dt.score(x_train, y_train)
    ts_acc = dt.score(x_test, y_test)
    
    # PLOT DECISION TREE BOUNDARY
    ax = plot_tree_boundary(x_train, y_train, dt, 'depth = ' + str(depth)\
                           + ', train acc = ' + str(tr_acc)\
                           + ', test acc = ' + str(ts_acc), ax)
    
    return ax
```

```python
#--------  plot_tree_boundary
# A function that visualizes the data and the decision boundaries
# Input: 
#      x (predictors)
#      y (labels)
#      model (the classifier you want to visualize)
#      title (title for plot)
#      ax (a set of axes to plot on)
# Returns: 
#      ax (axes with data and decision boundaries)

def plot_tree_boundary(x, y, model, title, ax):
    # PLOT DATA
    ax.scatter(x[y==1,0], x[y==1,1], c='green')
    ax.scatter(x[y==0,0], x[y==0,1], c='white')
    
    # CREATE MESH
    interval = np.arange(0,1,0.01)
    n = np.size(interval)
    x1, x2 = np.meshgrid(interval, interval)
    x1 = x1.reshape(-1,1)
    x2 = x2.reshape(-1,1)
    xx = np.concatenate((x1, x2), axis=1)

    # PREDICT ON MESH POINTS
    yy = model.predict(xx)    
    yy = yy.reshape((n, n))

    # PLOT DECISION SURFACE
    x1 = x1.reshape(n, n)
    x2 = x2.reshape(n, n)
    ax.contourf(x1, x2, yy, alpha=0.1, cmap='Greens')
    
    # LABEL AXIS, TITLE
    ax.set_title(title)
    ax.set_xlabel('Latitude')
    ax.set_ylabel('Longitude')
    
    return ax
```

## Challenge Problem: Meta Learning

In the problem, you are provided with 10 different previously trained prediction models for a spam classification task. The task is to investigate how can one combine these models into a single meta classification model (without retraining the individual models) that performs better than each of the individual ones? 

The data for this problem is provided in the files `dataset_5_train.txt` and `dataset_5_test.txt`. Each row of these files is an email described by 57 attributes, and the last column is 1 if the email is spam, and 0 otherwise. 

The prediction models are provided in the file `models.npy` and can be loaded into an array by executing: 

`models = np.load('models.npy')`

As before, you can make predictions using the $i^\text{th}$ using:

`model[i].predict(x_test)` 

and score the model using:

`model[i].score(x_test, y_test)`

The baseline for this task is a simple combination strategy that takes a majority vote from the individual prediction models.

**Any reasonable model that performs better than the baseline model on the test set will receive full credit.**

There are many intuitive ways to combine these 10 models into one; a more sophisticated approach is called "mixture of experts". In this problem, we are not requiring you to implement any particular approach.

```python
models = np.load('models.npy')

train = np.loadtxt('datasets/dataset_5_train.txt', delimiter=',')
test =  np.loadtxt('datasets/dataset_5_test.txt', delimiter=',')

x_test = test[:, 0:-1]
y_test = test[:, -1]

x_train = train[:, 0:-1]
y_train = train[:, -1]
```

## Baseline Model

```python
# iterate through models, generate predictions for the test set and take the mode. 

def find_mode(row):
    counts = np.unique(row, return_counts=True)[1]
    # resolve ties
    if (len(counts) == 2 and counts[0] == counts[1]):
        if np.random.random() >0.5:
            return counts[0]
        return counts[1]
    return np.argmax(counts)

# returns an n x 10 matrix of the predictions of each model on the testing set
def predict(x_test):
    predictions = np.zeros((x_test.shape[0], len(models)))
    for i in range(len(models)):
        predictions[:, i] = models[i].predict(x_test)
    return predictions

test_predictions = predict(x_test)
majority_vote = np.apply_along_axis( find_mode, axis=1, arr=test_predictions )
```

```python
baseline_accuracy = np.mean(majority_vote == y_test)
print 'baseline accuracy using a majority vote: ', baseline_accuracy
```

## Weighted voting

The 10 models can be combined in a number of ways. Two simple approaches involve a weighted average method where the predictions of each model are given a weight and then combined to generate a final prediction. 

### Option 1: Weight the predictions by the confidence score of the model. 

The confidence score of each model on the training set is used as the weight, and the predictions from each model are summed. The domain of y is changed from {0,1} to {-1, 1} so that the weights can be added and then averaged. If the average of the weights is greater than 0.5, the prediction is assumed to be a 1 (spam), and if the average of the weights is less than 0.5, the prediction is assumed to be a 0 (not spam). 

```python
prediction = np.zeros((len(y_test)))
for i in range(len(models)):
    val = models[i].predict(x_test)
    # set predictions of 0 to -1 so the weights can be summed
    val[val == 0] = -1
    prediction = prediction + models[i].score(x_train, y_train)*val

prediction = prediction/float(len(models))
```

```python
# rename response based on avergaed vote
prediction[prediction > 0] = 1
prediction[prediction < 0] = 0
```

```python
weighted_average = np.mean(prediction == y_test)
print 'weighted_average accuracy ', weighted_average
```

This weighted average model has resulted in an improvement in the accuracy of classification on the test set from 0.77 to 0.88 as some models are not as good as others and weighting, ensures that bad models contribute less to the overall prediction. 

### Option 2: Fit a classifier with the predictions from each of the given models as the features and use this to calculate the model weights

Another similar approach involves training a classifier (such as logistic regression) on the predictions from the models - i.e. the predictors are the outputs of each of the 10 models (resulting in an n x 10 predictor matrix) and the response is the y variable indicating if the email is spam or not. The logistic regression model will calculate the coefficients for each of the predictors which are essentially the weights given to the predictions generated by each model. 

```python
# logisitc regression classifier
train_predictions = predict(x_train)
log_model = LogisticRegression().fit(train_predictions, y_train)
test_predict_y = log_model.predict(test_predictions)
log_model_accuracy = np.mean(test_predict_y == y_test)
print 'log_model_accuracy  ', log_model_accuracy
```

```python
# random forest classifier
tree_model = RandomForestClassifier(n_estimators=200).fit(train_predictions, y_train)
test_predict_y_tree = tree_model.predict(test_predictions)
tree_model_accuracy = np.mean(test_predict_y_tree == y_test)
print 'tree_model_accuracy  ', tree_model_accuracy
```

These classification models trained with the output of the model predictions results in a further improvement in the accuracy of the predictions with the accuracy increasing to 0.897. 

These approaches reduce the effect of a noisy/inaccurate model on the ensemble prediction by weighting the predictions from each model appropriately.
