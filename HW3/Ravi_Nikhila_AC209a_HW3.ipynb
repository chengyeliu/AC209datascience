### Part (a): Implement multiple linear regression from scratch

You are provided a data set containing attributes related to automobiles as well as their corresponding prices. The task is to build a linear regression model from scratch that can estimate the price of an automobile (response variable) using its attributes (predictor variables).

The file ``dataset_1_train.txt`` contains the training set that you can use to fit a regression model, and the file ``dataset_1_test.txt`` contains the test set that you can use to evaluate the model. In each file, the first two columns contain the predictors of the automobile, namely ``'horsepower'`` and ``'highway MPG'``, and the last column contains the automobile prices.

- Implement the following two functions from scratch. 

    - ``multiple_linear_regression_fit``:

        - takes as input: the training set, ``x_train``, ``y_train``

        - fits a multiple linear regression model

        - returns the model parameters (coefficients on the predictors, as an array, and the intercept, as a float).

    - ``multiple_linear_regression_score``:

        - takes model parameters (coefficients and intercept) and the test set, ``x_test`` ``y_test``, as inputs

        - returns the R^2 score for the model on the test set, along with the predicted y-values.
        
- Use your functions to predict automobile prices and evaluate your predictions.

**Note:** You **may not** use pre-built models or model evaluators for these tasks.

```python
### Functions for fitting and evaluating multiple linear regression

#--------  multiple_linear_regression_fit
# A function for fitting a multiple linear regression
# Fitted model: f(x) = x.w + c
# Input: 
#      x_train (n x d array of predictors in training data)
#      y_train (n x 1 array of response variable vals in training data)
# Return: 
#      w (d x 1 array of coefficients) 
#      c (float representing intercept)

def multiple_linear_regression_fit(x_train, y_train):
    
    # Append a column of one's to x
    n = x_train.shape[0]
    ones_col = np.ones((n, 1))
    
    x_train = np.concatenate((x_train, ones_col), axis=1)
    
    # Compute transpose of x
    x_transpose = np.transpose(x_train)
    
    # Compute coefficients: w = inv(x^T * x) x^T * y
    # Compute intermediate term: inv(x^T * x)
    # Note: We have to take pseudo-inverse (pinv), just in case x^T * x is not invertible 
    x_t_x_inv = np.linalg.pinv(np.dot(x_transpose, x_train))
    
    # Compute w: inter_term * x^T * y 
    w = np.dot(np.dot(x_t_x_inv, x_transpose), y_train)
    
    # Obtain intercept: 'c' (last index)
    c = w[-1]
    
    return w[:-1], c

#--------  multiple_linear_regression_score
# A function for evaluating R^2 score and MSE 
# of the linear regression model on a data set
# Input: 
#      w (d x 1 array of coefficients)
#      c (float representing intercept)
#      x_test (n x d array of predictors in testing data)
#      y_test (n x 1 array of response variable vals in testing data)
# Return: 
#      r_squared (float) 
#      y_pred (n x 1 array of predicted y-vals)

def multiple_linear_regression_score(w, c, x_test, y_test):        
    # Compute predicted labels
    y_pred = np.dot(x_test, w) + c
    
    # Evaluate sqaured error, against target labels
    # sq_error = \sum_i (y[i] - y_pred[i])^2
    sq_error = np.sum(np.square(y_test - y_pred))
    
    # Evaluate squared error for a predicting the mean value, against target labels
    # variance = \sum_i (y[i] - y_mean)^2
    y_mean = np.mean(y_test)
    y_variance = np.sum(np.square(y_test - y_mean))
    
    # Evaluate R^2 score value
    r_squared = 1 - sq_error / y_variance

    return r_squared, y_pred
```

```python
# Load train and test data sets
data_train = np.loadtxt('datasets/dataset_1_train.txt', delimiter=',', skiprows=1)
data_test = np.loadtxt('datasets/dataset_1_test.txt', delimiter=',', skiprows=1)

# Split predictors from response
# Training
y_train = data_train[:, -1]
x_train = data_train[:, :-1]

# Testing
y_test = data_test[:, -1]
x_test = data_test[:, :-1]

# Fit multiple linear regression model
w, c = multiple_linear_regression_fit(x_train, y_train)

# Evaluate model
r_squared, _ = multiple_linear_regression_score(w, c, x_test, y_test)

print 'R^2 score on test set:', r_squared 
```

## Problem 2: Polynomial regression
In this problem, we revisit a dataset from Homework 1 and fit polynomial regression models to it. The dataset is provided in the file ``dataset_3.txt``, which contains a single predictor variable ``x`` in the first column and the response variable ``y`` in the second column. 

### Part(a): Implement polynomial regression from scratch

- Implement the following three functions from scratch:

    - ``polynomial_regression_fit``:
        - takes as input: training set, ``x_train``, ``y_train`` and the degree of the polynomial
        - fits a polynomial regression model 
        - returns the model parameters (array of coefficients and the intercept)

    - ``polynomial_regression_predict``: 
        - takes as input: the model parameters (array of coefficients and the intercept), the degree of the polynomial and the test set predictors ``x_test``
        - returns the response values predicted by the model on the test set. 

    - ``polynomial_regression_score``: 
        - takes an array of predicted response values and the array of true response values ``y_test``
        - returns R^2 score for the model on the test set, as well as the sum of squared errors

- Fit polynomial regression models of degrees 3, 5, 10 and 25 to the data set. Visualize the original data along with the fitted models for the various degrees in the same plot. 

For this problem, you may either use the multiple linear regression functions implemented in the Problem 1 or use the in-built functions in ``sklearn``.

```python
def polynomial_regression_fit(x_train, y_train, degrees):
    # Create the poly terms for x,x^2 .. 
    n= np.size(x_train)   # data size 
    x_poly = np.zeros([n, degrees]) # poly matrix
    for d in range(1, degrees +1):
        x_poly[:, d - 1] = np.power(x_train.transpose()[0], d) 
    # Use the poly matrix as the input to multiple linear regression - each column of the matrix is 
    # treated as a predictor i.e. x, x^2
    w,c  = multiple_linear_regression_fit(x_poly, y_train)
    return w, c

def polynomial_regression_predict(w,c,x,degrees):
    # array of params
    wc = np.append(w,c)
    n= x.shape[0]  # data size 
    x_poly = np.zeros([n, degrees]) # poly degree 
    # reconstruct the polynomial matrix of x values 
    for d in range(1, degrees +1):
        x_poly[:, d - 1] = np.power(x.transpose()[0], d) 
    # append column of ones to multiply the x intercept
    ones_col = np.ones((n,1))
    x_poly = np.concatenate((x_poly, ones_col), axis=1)
    y_pred = np.dot(wc, x_poly.T)
    return y_pred

def polynomial_regression_score(y_pred, y_test):
    RSS = np.sum((y_pred-y_test)**2)
    # mean of y
    y_bar = np.mean(y_test)
    TSS = np.sum((y_bar - y_test)**2)
    R2 = 1 - RSS/TSS
    return R2

# Load train set
data = np.loadtxt("datasets/dataset_3.txt", delimiter=',', skiprows = 1)
data = np.array(sorted(data, key=lambda elem: elem[0]))
X_test = data[:, :-1]
Y_test = data[:, -1]

def plot_pred(X_test,Y_test, degrees, color):
    w, c = polynomial_regression_fit(X_test,Y_test, degrees)
    y_pred = polynomial_regression_predict(w,c,X_test,degrees)
    plt.plot(X_test,y_pred, color=color, label='poly '+str(degrees) + ' degree')

plt.figure(figsize=(10,10))
plt.plot(X_test,Y_test, 'ro', label='Test Data')
plt.xlabel('X'); plt.ylabel('Y')
plot_pred(X_test,Y_test, 3, 'blue')
plot_pred(X_test,Y_test, 5, 'green')
plot_pred(X_test,Y_test, 10, 'yellow')
plot_pred(X_test,Y_test, 25, 'pink')
plt.legend(loc='best')
```

The 3 and 5 degree polynomials appear to be good fits to the data capturing the peak and trough.  The 25 degree polynomial shows signs of overfitting with excess curvature and so may not generalise well to other test sets. 

###  Part (b): Comparing training and test errors

- Split the data set in Problem 2 each into training and test sets: use the first 50% of the data for training and the remaining for testing. 


- Fit polynomial models of varying degree ranging from 1 to 15 to the training sets. Evaluate  the various fits on **both** the training and the test sets. Plot both the R^2 score of the fitted polynomial models on the training and test sets as a functions of the degree. 


- Describe the relationship between degree of the polynomial model and the fit on both the training and testing data. Explain, based on the plot, what is the best polynomial model for the data.

```python
# load the data
data = np.loadtxt("datasets/dataset_3.txt", delimiter=',', skiprows = 1)
n = data.shape[0]
# shuffle the data
samples = np.random.permutation(n)
# split into testing and training sets
Xtest = data[samples[:50], :-1]
Ytest = data[samples[:50], 1]
Xtrain = data[samples[50:], :-1]
Ytrain = data[samples[50:], 1]

def AIC_score(n,k,RSS):
    score = (n*np.log(RSS/float(n)) + 2*k)
    return score

def BIC_score(n,k,RSS):
    score = (n*np.log(RSS/float(n)) + np.log(n)*k)
    return score

R2_train = np.zeros((15,1))
R2_test = np.zeros((15,1))
AIC_train = np.zeros((15,1))
BIC_train = np.zeros((15,1))

degrees = np.array(range(1,16)).reshape(15,1)

for d in degrees:
    # fit a polynomial of degree d and obtain the params
    w, c = polynomial_regression_fit(Xtrain,Ytrain,d)
    
    # use the params to calculate predictions for the training and testing set
    Ytrain_pred = polynomial_regression_predict(w,c, Xtrain, d)
    Ytest_pred = polynomial_regression_predict(w,c, Xtest, d)
    
    # calculate R2 for test and train sets
    r2_train = polynomial_regression_score(Ytrain_pred, Ytrain)
    r2_test = polynomial_regression_score(Ytest_pred, Ytest)
    
    # calculate AIC BIC for the training set
    RSS_train = np.sum((Ytrain_pred-Ytrain)**2)
    aic_train = AIC_score(50,d, RSS_train)
    bic_train = BIC_score(50,d, RSS_train)
    
    # save values
    R2_train[d-1] = r2_train
    R2_test[d-1] = r2_test
    AIC_train[d-1] = aic_train
    BIC_train[d-1] = bic_train

plt.figure(figsize=(10,5))
plt.plot(degrees, R2_train, 'bo-', label='R2 Test')
plt.plot(degrees, R2_test, 'ro-', label='R2 Train')
plt.legend(loc=4)
plt.xlabel('Polynomial regression degree')
plt.ylabel('R2 value of fit')
plt.legend(loc='best')
```

**Describe the relationship between degree of the polynomial model and the fit on both the training and testing data. Explain, based on the plot, what is the best polynomial model for the data.**

The $R^2$ value increases as the polynomial degree increases. As more terms are added to the polynomial, the model becomes a closer fit to the training data so the RSS decreases which causes the $R^2$ to increase. 

The largest increase in $R^2$ results when increasing the polynomial degree from 2 to 3. For degree 3, the difference in $R^2$ value between testing and training sets is also the lowest out of all the polynomial degrees tested. There are only very small increases in $R^2$ value for polynomials with degrees higher than 3. 

The $R^2$ value for the testing set also starts to decrease for polynomials greater than 3 as these models are likely to be overfitting to the training data and so not generalise well to the test data. 

Therefore the best polynomial for this data is of degreee 3.

### Part (b): Application to New York taxi cab density estimation

We shall now apply the concepts learned so far to a real-world prediction task. You are asked to build a regression model for estimating the density of Green cab taxis at any given time of a day in New York city. The model needs to take the time of the day (in minutes) as input, and predict the expected number of pick ups at that time.

The data set for this problem can be downloaded from the following URL: https://s3.amazonaws.com/nyc-tlc/trip+data/green_tripdata_2015-01.csv. The file contains the details of all pickups by Green cabs in New York City during January 2015. 

```python
df = pd.read_csv('./datasets/green_tripdata_2015-01.csv', skiprows=[0], usecols=[1], names=['pick-up datetime'])
```

```python
df.head()
```

```python
# add a column of datetime objects 
df.rename(columns={'pick-up datetime':'times'}, inplace=True)
df['times'] = pd.to_datetime(df['times'])
# for each time, calculate the total number of elapsed minutes normalised by the total number of minutes in a day 
df['pickup_time'] = df['times'].apply(lambda row: (row.hour*60 + row.minute)/float(1440))

# count the number of pickups at that time across the month  
times = df.groupby('pickup_time', as_index=False).count()
# divide by the number of days in a month to get the average number of pickups at that minute
times['average_pickups'] = times['times']/float(31)
times.head()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(times['pickup_time'].values, times['average_pickups'].values)
plt.xlabel('normalised time in minutes during the day')
plt.ylabel('average number of pickups during January')
```

```python
# Create fitting functions that use the statsmodel api 

def polynomial_regression_fit(x, y, degrees):
    # Create the poly terms for x,x^2 .. 
    
    n= np.size(y)   # data size 
    x_poly = np.zeros([n, degrees]) # poly degree 

    for d in range(1, degrees +1):
        x_poly[:, d - 1] = np.power(x, d)  # adding terms 

    Xt=sm.add_constant(x_poly)
    model=sm.OLS(y,Xt)
    model_results=model.fit()
    return model_results, Xt

def polynomial_regression_predict(params, degrees, x):
    # # Create the poly terms for x,x^2 ..
    n = x.shape[0]
    x_poly = np.zeros([n, degrees])
    for d in range(1, degrees + 1):
        x_poly[:, d - 1] = np.power(x, d)
    Xt=sm.add_constant(x_poly)
   
    # Predict y-vals
    y_pred = np.dot(params,Xt.T)
        
    return y_pred
```

```python
# split the data into testing and training sets
data_train = times.sample(frac=0.7)
data_test = times.drop(data_train.index)

Xtest = data_test['pickup_time'].values
Ytest = data_test['average_pickups'].values
Xtrain = data_train['pickup_time'].values
Ytrain = data_train['average_pickups'].values

# set the range of degrees of polynomials to test
degrees=30
degree_range = range(1, degrees +1)

# create empty lists to store R2 and AIC/BIC values
R2_train = np.zeros([ degrees+1,1])
R2_test = np.zeros([ degrees+1,1])
AIC = np.zeros([ degrees+1,1])
BIC = np.zeros([ degrees+1,1])

# fit the model using different polynomial degrees
for d in degree_range:
    # fit the model
    model_results, _= polynomial_regression_fit(Xtrain,Ytrain, d)
    params = model_results.params
    
    # use the model to predict the y values on the test set
    y_test_pred = polynomial_regression_predict(params, d, Xtest)
    
    # calculate the R2 values for testing and training sets and AIC and BIC for training
    R2_train[d] = model_results.rsquared
    R2_test[d] = polynomial_regression_score(y_test_pred, Ytest)
    AIC[d] = model_results.aic
    BIC[d] = model_results.bic

plt.figure(figsize=(10,5))
plt.plot(degree_range, R2_train[1:], 'bo-', label='R2 Train')
plt.plot(degree_range, R2_test[1:], 'ro-', label='R2 Test')
plt.xlabel('degree of polynomial'); 
plt.ylabel(r'$R^2$')
plt.legend(loc='best')

plt.figure(figsize=(10,5))
plt.plot(degree_range, AIC[1:], 'bo-', label='AIC Train')
plt.plot(degree_range, BIC[1:], 'ro-', label='BIC Train')
plt.xlabel('degree of polynomial'); 
plt.ylabel('AIC/BIC')
plt.legend(loc='best')
```

```python
plt.figure(figsize=(15,10))
def plot_model(d, color):
    model_results, _= polynomial_regression_fit(Xtrain,Ytrain, d)
    params = model_results.params
    y_test_pred = polynomial_regression_predict(params, d, Xtest)
    plt.plot(Xtest, y_test_pred, linewidth=1.5, color=color, label=str(d) + ' degree model')
    
plt.scatter(Xtest, Ytest, label='Test Data')
plot_model(3, 'red')
plot_model(4, 'green')
plot_model(7, 'yellow')
plot_model(10, 'pink')
plot_model(15, 'orange')
plot_model(20, 'magenta')
plt.xlabel('normalised time in minutes during the day')
plt.ylabel('average number of pickups during January')
plt.legend(loc='best')
```

The values of $\beta$ can be calculated in the same way as for polynomial regression, by adding all the polynomial terms to the x matrix used in the least squares calculation. 


```python
df_train = np.loadtxt('datasets/dataset_1_train.txt', skiprows=1, delimiter=',')
df_test = np.loadtxt('datasets/dataset_1_test.txt', skiprows=1, delimiter=',')

x_train = df_train[:,:-1]
y_train = df_train[:,-1]
x_test = df_test[:,:-1]
y_test = df_test[:,-1]

x1 = x_test[:,0]
x2 = x_test[:,1]

def create_2d_x_poly(x):
    n= x.shape[0]   # data size 
    x1 = x[:,0]
    x2 = x[:,1]
    x_poly = np.zeros([n, 5]) # poly degree 
    x_poly[:, 0] = x1
    x_poly[:, 1] = x2
    x_poly[:, 2] = np.power(x1, 2)
    x_poly[:, 3] = x1*x2
    x_poly[:, 4] = np.power(x2, 2)
    return x_poly

def create_3d_x_poly(x):
    n= x.shape[0]   # data size 
    x1 = x[:,0]
    x2 = x[:,1]
    x_poly = np.zeros([n,9])
    x_poly[:, 0] = x1
    x_poly[:, 1] = x2
    x_poly[:, 2] = np.power(x1, 2)
    x_poly[:, 3] = x1*x2
    x_poly[:, 4] = np.power(x2, 2)
    x_poly[:, 5] = np.power(x1, 3)
    x_poly[:, 6] = np.power(x1, 2)*x2
    x_poly[:, 7] = np.power(x2, 2)*x1
    x_poly[:, 8] = np.power(x2, 3)
    return x_poly

def polynomial_regression_predict_2deg(params, x_test):
    x_poly = create_2d_x_poly(x_test)
    Xt=sm.add_constant(x_poly)
    # Predict y-vals
    y_pred = np.dot(params,Xt.T)
    return y_pred

def multivariate_polynomial_fit_2deg(x_train, y_train):
    # Create the poly terms for x1,x1^2, x2, x2^2 .. 
    x_poly = create_2d_x_poly(x_train)
    Xt=sm.add_constant(x_poly)
    model=sm.OLS(y_train,Xt)
    model_results=model.fit()
    return model_results

def polynomial_regression_predict_3deg(params, x_test):
    x_poly = create_3d_x_poly(x_test)
    Xt=sm.add_constant(x_poly)
    y_pred = np.dot(params,Xt.T)
    return y_pred

def multivariate_polynomial_fit_3deg(x_train, y_train): 
    x_poly = create_3d_x_poly(x_train)
    Xt=sm.add_constant(x_poly)
    model=sm.OLS(y_train,Xt)
    model_results=model.fit()
    return model_results

def plot(y_pred):
    fig = plt.figure(figsize=(12, 8))
    ax = Axes3D(fig, azim=-150, elev=10)
    ax.scatter(x1, x2, y_test, label='test data')
    ax.scatter(x1, x2, y_pred, color='red', label='predictions')
    plt.legend(loc='best')
    ax.set_xlabel('horsepower')
    ax.set_ylabel('highway MPG')
    ax.set_zlabel('Price')

# fit to a 2 degree polynomial 
model_results = multivariate_polynomial_fit_2deg(x_train, y_train)
y_test_pred = polynomial_regression_predict_2deg(model_results.params, x_test)
R2 = polynomial_regression_score(y_test_pred, y_test)
plot(y_test_pred)
print 'R2 for 2 degree multivariate polynomial', R2
print 'AIC 2 degrees:', model_results.aic
print 'BIC 2 degrees:', model_results.bic

# fit to a 3 degree polynomial

model_results = multivariate_polynomial_fit_3deg(x_train, y_train)
y_test_pred = polynomial_regression_predict_3deg(model_results.params, x_test)
R2_3deg = polynomial_regression_score(y_test_pred, y_test)
plot(y_test_pred)
print 'R2 for 3 degree multivariate polynomial', R2_3deg
print 'AIC 3 degrees:', model_results.aic
print 'BIC 3 degrees:', model_results.bic
```

### Part (b): Weighted linear regression

Suppose you are told that some of the prices recorded in the training set are noisy, and you are given the list of noisy points, how would you use this information during training to fit a better regression model?

The noise level for each training point is provided in the file dataset_1_train_noise_levels.txt. A noise level 'none' indicates that the price is accurate, and a noise level 'noisy' indicates that the price is only moderately accurate. 

We want to fit a linear regression model that accounts for this new information. One way to do this is to assign different weights to each training point based on the amount of noise associated to that training point. That is, our loss function is now
$$
\sum_{i=1}^n \alpha_i\,(y_i \,-\, w^T x_i)^2
$$
where $\alpha_i$ is a number representing how much you value the contribution of the data point $x_i$.

How does the $R^2 score$ (evaluated on the test set) of the new linear model compare to the one fitted using plain linear regression?

**ANSWER**

The expression of the coefficients of the linear regression model now becomes: 

$$ \hat{\beta} = {\bf (X^{T}AX)^{-1} X^T AY } $$

where A is a diagonal weight matrix. 

This can be used in the least squares calculation to computer the coefficients of the polynomial. 

```python
noise_levels = np.loadtxt('datasets/dataset_1_train_noise_levels.txt', skiprows=1, dtype='string')

# start with a weight of 0.5 for noisy data and 1 for non noisy data 
noise = [0.5 if (noise == 'noisy') else 1 for noise in noise_levels]

# create a diagnoal weight matrix
A = np.diag(noise)

# function that computes the weights using the expression given above 
def weighted_linear_regression_fit(x_train, y_train, A):
    n = x_train.shape[0]
    m = x_train.shape[1]
    ones_col = np.ones((n,1))
    
    # add a column of ones to mulitply the constant term 
    x = np.concatenate((x_train, ones_col), axis=1)
    xT = np.transpose(x)
    
    # intermediate term xTA
    xTA = np.dot(xT,A)
    xTAx = np.dot(xTA,x)
    xTAx_invXT = np.dot(np.linalg.pinv(xTAx), xT)
    
    # final matrix 
    xTAx_invXTA = np.dot(xTAx_invXT, A)
    
    # dot the final matrix with y_train
    w = np.dot(xTAx_invXTA, y_train)
    c = w[-1]
    return w[:-1], c

def weighted_linear_regression_score(w, c, x_test, y_test):
    y_pred = np.dot(x_test,w) + c
    RSS = np.sum((y_pred-y_test)**2)
    y_bar = np.mean(y_test)
    TSS = np.sum((y_bar - y_test)**2)
    R2 = 1 - RSS/TSS
    return R2, y_pred

w,c = weighted_linear_regression_fit(x_train, y_train, A)
R2, y_pred = weighted_linear_regression_score(w,c, x_test, y_test)

x1 = x_test[:,0]
x2 = x_test[:,1]
xx1, xx2 = np.meshgrid(np.linspace(x1.min(), x1.max(), 100), 
                       np.linspace(x2.min(), x2.max(), 100))

Z = c + w[0] * xx1 + w[1] * xx2
fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-140, elev=1)
surf = ax.plot_surface(xx1, xx2, Z, alpha=0.3, linewidth=0)
ax.scatter(x1, x2, y_test, label='test data')
ax.scatter(x1, x2, y_pred, color='red', label='predictions')
plt.legend(loc='best')
ax.set_xlabel('horsepower')
ax.set_ylabel('highway MPG')
ax.set_zlabel('Price')
```

```python
print 'R2 value for weighted linear regression: ', R2
```

The R2 value has improved from 0.177 for simple linear regression to 0.341 using weighted linear regression, indicating that dampening the noise terms has improved the prediction. Reducing the effect of the noisy terms, means the model can fit to the 'clean' data and does not need to capture the behaviour of the noisy terms.
